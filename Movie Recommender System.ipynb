# Recommender Systems


## Data Processing and Visualization Part

```python
import numpy as np
import pandas as pd
```

```python
columns_names = ['user_id','item_id','rating','timestamp']
```

```python
df = pd.read_csv('u.data',sep = '\t',names = columns_names)

df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
df = pd.merge (df,movie_titles,on='item_id')
df.head()
```

```python
# Let's see the highest rated films by the title in top 5 order.
df.groupby('title')['rating'].mean().sort_values(ascending = False).head()
```

```python
# Let's see the movies that has the highest vote count by the title in top 5 order.
df.groupby('title')['rating'].count().sort_values(ascending = False).head()
```

```python
# Created a datafrome for the ratings by film. Used the mean value for the ratings.
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
# Added a vote counter to the dataframe.
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.figure(figsize = (13,9))
sns.set_style('whitegrid')
# Number of the ratings from the movies 
sns.distplot(ratings['num of ratings'] , kde=False, bins =70)
```

```python
# Actual ratings of the movies

plt.figure(figsize = (13,9))
sns.set_style('whitegrid')
sns.distplot(ratings['rating'] , kde=False, bins =70)
```

```python
sns.set_style('whitegrid')

sns.jointplot('rating','num of ratings', data = ratings, alpha = 0.5,height = 11)
```

# Recommendation Part 

```python
# Created a table for our movie titles by individual user ratings.
movimat = df.pivot_table(index = 'user_id',columns='title',values = 'rating')

movimat.head()
```

```python
# Then sorted the vote counts in descending format
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
# Then we selected some movies for practice.
starwars_user_ratings = movimat['Star Wars (1977)']
liarliar_user_ratings = movimat['Liar Liar (1997)']
```

```python
# Then we checked the individual user ratings of the selected movie.
starwars_user_ratings.head()
```

```python
# Then we checked that if he have some correlation between the user ratings of that film and other films.
similar_to_starwars = movimat.corrwith(starwars_user_ratings)
```

```python
# Same correlation relation with this movie also.
similar_to_liarliar = movimat.corrwith(liarliar_user_ratings)
```

```python
# Then we created a correlation dataframe from the similar results of the first film. Dropped NaN values from that database.
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace = True)

corr_starwars.head()
```

```python
# Let's check out the sorted correlation values.
corr_starwars.sort_values('Correlation',ascending = False).head(10)
```

```python
# As we can see there are a lot of high rated but insufficient voted films there. So we should filter out those less voted films.

corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings'] > 100].sort_values('Correlation',ascending = False).head()
```

```python
# Let's do the samething for the Film Liar Liar.
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns = ['Correlation'])

corr_liarliar.dropna(inplace = True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])

corr_liarliar[corr_liarliar['num of ratings'] > 100].sort_values('Correlation',ascending = False).head()
```
